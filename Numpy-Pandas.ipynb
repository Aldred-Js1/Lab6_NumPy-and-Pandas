```python
import pandas as pd

# Step 1: Create a NumPy array of random data 
data = np.random.randint(1, 50, size=(5, 3)) 

# Step 2: Convert the array into a DataFrame 
df_data = pd.DataFrame(data, columns=['X', 'Y', 'Z']) 
print("Generated Data:\n", df_data) 

# Step 3: Apply NumPy functions to DataFrame columns 
df_data['Log_X'] = np.log(df_data['X']) 
df_data['Sqrt_Y'] = np.sqrt(df_data['Y']) 
print("\nDataFrame with NumPy Transformations:\n", df_data) 

# Step 4: Analyze data using Pandas correlation 
print("\nCorrelation Matrix:\n", df_data.corr())

# Try it:
# - Use np.mean() and np.median() to compute overall column statistics
# - Apply other NumPy functions such as np.square() or np.exp()

# Compute Mean and Median for Column 'X'
mean_x = np.mean(df_data['X'])
median_x = np.median(df_data['X'])
overall_mean = np.mean(df_data, axis=0) 
print(f"\nColumn X - Mean: {mean_x:.2f}, Median: {median_x}")
print("\nOverall Column Means:\n", overall_mean)

# Square values in Column Z
df_data['Z_Squared'] = np.square(df_data['Z'])
# Exponential transformation to Column X
df_data['Exp_X'] = np.exp(df_data['X'])
print("\nUpdated DataFrame with Square and Exp:\n", df_data[['X', 'Z', 'Z_Squared', 'Exp_X']].head())
```

```python
# Step 1: Save the DataFrame to a CSV file 
df_data.to_csv('sample_data.csv', index=False) 
print("Data saved to 'sample_data.csv'.") 

# Step 2: Load the CSV file back into a new DataFrame 
df_imported = pd.read_csv('sample_data.csv') 
print("Imported DataFrame:\n", df_imported) 

# Step 3: Generate summary statistics 
summary_stats = df_imported.describe() 
print("Summary Statistics:\n", summary_stats) 

# Step 4: Display column means and standard deviations 
print("Column Means:\n", df_imported.mean()) 
print("Column Standard Deviations:\n", df_imported.std())

# Try it:
# Add a new calculated column (e.g., df_imported['Sum_XY'] = df_imported['X'] + df_imported['Y'])
df_imported['Sum_XY'] = df_imported['X'] + df_imported['Y']
print("\nDataFrame with Sum_XY Column:\n", df_imported[['X', 'Y', 'Sum_XY']].head())
# Save the modified DataFrame to a new CSV file and review
df_imported.to_csv('modified_sample_data.csv', index=False)
print("Modified data saved to 'modified_sample_data.csv'.")
```

```python
# Step 1: Load the dataset
# Added low_memory=False to remove the warning you saw
df_kaggle = pd.read_csv('survey_results_public.csv', low_memory=False) 

# Step 2: Select the correct columns based on your file's actual names
# We are using 'WorkExp' instead of 'YearsCodePro' and 'ConvertedCompYearly' for salary
df_subset = df_kaggle[['Country', 'EdLevel', 'WorkExp', 'ConvertedCompYearly']].copy()

# Step 3: Clean the data
df_clean = df_subset.dropna().copy()

# Step 4: Categorize experience into groups 
# Since WorkExp is already a number, we don't need to convert it!
df_clean['ExperienceLevel'] = np.where(df_clean['WorkExp'] >= 10, 'Senior', 'Junior') 

# Step 5: Group data by Country and ExperienceLevel, compute average salary 
# Note: Using ConvertedCompYearly here
grouped_data = df_clean.groupby(['Country', 'ExperienceLevel'])['ConvertedCompYearly'].mean()

# Step 6: Reset index for readability 
grouped_data = grouped_data.reset_index() 

print("--- Success! Formatted Grouped Data ---")
print(grouped_data.head(10))

# Try it: 
import pandas as pd

# --- Step 1: Education Level Analysis ---
# We calculate mean, median, and standard deviation in one clean command
ed_stats = df_clean.groupby(['Country', 'EdLevel'])['ConvertedCompYearly'].agg(['mean', 'median', 'std']).reset_index()

# Rename for professional presentation
ed_stats.columns = ['Country', 'Education_Level', 'Avg_Salary', 'Median_Salary', 'Std_Dev']

# Round the numbers to 2 decimal places so they look like currency
ed_stats = ed_stats.round(2)

print("\n--- Salary Statistics by Education Level (First 10) ---")
print(ed_stats.head(10))


# --- Step 2: High-Earner Analysis ---
# Find the top 10 countries based on average compensation
top_countries = ed_stats.sort_values(by='Avg_Salary', ascending=False).head(10)

print("\n--- Top 10 Countries with Highest Average Compensation ---")
# to_string(index=False) makes it look like a clean report instead of a raw dataframe
print(top_countries[['Country', 'Education_Level', 'Avg_Salary']].to_string(index=False))
```
